```python
import pandas as pd
import matplotlib.pyplot as plt

link = 'https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv'
code = '?raw=true'
url = link + code
df = pd.read_csv(url, sep=',', index_col=1)  # Corregir el Ã­ndice a 0
```

```python
df
```

```python
df_usa = df[df['Country'] == 'United States of America']
```

```python
print(df.columns)
```

```python
print(df_usa)
```

```python

mean_values=df_usa['No. of confirmed cases'].mean()
```

```python
df_usa.fillna(mean_values,inplace=True)
```

```python
USA=df_usa[10:25]
```

```python
print(df_usa.columns)
```

```python
Dates=df_usa.index[0:-1:10]##################VARIABLE UNO
NCC=df_usa['No. of confirmed cases'][0:-1:10]#crea una variable para los NCC, del primer dato hasta el ultimo de 10 en 10
NCD=df_usa['No. of confirmed deaths'][0:-1:10]#""""
```

```python
Dates=df_usa.index[0:-1:10]#########################VAIABLE DOS
Male_cases=df_usa['Male Cases'][0:-1:10]#crea una variable para los male_cases, del primer dato hasta el ultimo de 10 en 10
Female_cases=df_usa['Female Cases'][0:-1:10]#""""
```

```python
Dates=df_usa.index[0:-1:10]########################VARIABLE TRES
Boys=df_usa['Boys'][0:-1:10]#crea una variable para los boys, del primer dato hasta el ultimo de 10 en 10
Girls=df_usa['Girls'][0:-1:10]#""""
```

```python
Dates=df_usa.index[0:-1:10]########################VARIABLE CUATRO
Adult_male=df_usa['Adult Male'][0:-1:10]#crea una variable para los adult male, del primer dato hasta el ultimo de 10 en 10
Adult_female=df_usa['Adult Female'][0:-1:10]#""""
```

```python
Dates=df_usa.index[0:-1:10]##########################VARIABLE CINCO
Old_male=df_usa['Old Male'][0:-1:10]#crea una variable para los old male, del primer dato hasta el ultimo de 10 en 10
Old_female=df_usa['Old Female'][0:-1:10]#""""
```

```python
#####################################################GRAFICA UNO
fig=plt.figure(figsize=(12,10))
plt.bar(Dates,NCC,color='red',alpha=0.2)
plt.plot(Dates,NCC,'o--m',ms=4,label='Infectados')
plt.plot(Dates,NCD,'d--b',ms=4,label='Decesos')
plt.bar(Dates,NCD)
plt.grid()
plt.title('Avance del ebola en Estados Unidos')#titulo
plt.xlabel('Periodo de tiempo')#etiqueta eje x
plt.ylabel('Numero de casos confirmados')#etiqueta en eje y
plt.xticks(rotation=90,fontsize=8)
plt.legend()#cuadro de info
plt.savefig("barras_ncc,nmc.png",dpi=300, format='png')
plt.show()
```

```python
##############################GRAFICA DOS
fig=plt.figure(figsize=(12,10))
plt.bar(Dates,Male_cases,color='red',alpha=0.2)
plt.plot(Dates,Male_cases,'o--m',ms=4,label='Infectados Masculinos')
plt.plot(Dates,Female_cases,'d--b',ms=4,label='Infectads Femeninos')
plt.bar(Dates,Female_cases)
plt.grid()
plt.title('Casos Masculinos y Femeninos')#titulo
plt.xlabel('Periodo de tiempo')#etiqueta eje x
plt.ylabel('Numero de casos confirmados')#etiqueta en eje y
plt.xticks(rotation=90,fontsize=8)
plt.legend()#cuadro de info
plt.savefig("barras_male,female.png",dpi=300, format='png')
plt.show()
```

```python
###################################Grafica tres
fig=plt.figure(figsize=(12,10))
plt.bar(Dates,Boys,color='red',alpha=0.2)
plt.plot(Dates,Boys,'o--m',ms=4,label='Hombres jovenes')
plt.plot(Dates,Girls,'d--b',ms=4,label='Mujeres jovenes')
plt.bar(Dates,Girls)
plt.grid()
plt.title('Casos Masculinos jovenes y Femeninos jovenes')#titulo
plt.xlabel('Periodo de tiempo')#etiqueta eje x
plt.ylabel('Numero de casos ')#etiqueta en eje y
plt.xticks(rotation=90,fontsize=8)
plt.legend()#cuadro de info
plt.savefig("barras_boys,gilrs.png",dpi=300, format='png')
plt.show()
```
