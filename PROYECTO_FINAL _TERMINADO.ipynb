```python
# Generar variaciones aleatorias enteras y positivas
variation_cases = np.random.randint(low=0, high=7, size=len(df_usa))
variation_deaths = np.random.randint(low=0, high=4, size=len(df_usa))

# Agregar variaciones a los datos existentes
df_usa['No. of confirmed cases'] += variation_cases
df_usa['No. of confirmed deaths'] += variation_deaths
```

```python

# Generar variaciones aleatorias enteras y positivas para 'Male Cases'
variation_male_cases = np.random.randint(low=0, high=10, size=len(df_usa))

# Generar variaciones aleatorias enteras y positivas para 'Female Cases'
variation_female_cases = np.random.randint(low=0, high=6, size=len(df_usa))

# Agregar variaciones a los datos existentes
df_usa['Male Cases'] += variation_male_cases
df_usa['Female Cases'] += variation_female_cases
```

```python
# Generar variaciones aleatorias enteras y positivas para 'Boys'
variation_boys = np.random.randint(low=0, high=5, size=len(df_usa))

# Generar variaciones aleatorias enteras y positivas para 'Girls'
variation_girls = np.random.randint(low=0, high=3, size=len(df_usa))

# Agregar variaciones a los datos existentes
df_usa['Boys'] += variation_boys
df_usa['Girls'] += variation_girls
```

```python
# Generar variaciones aleatorias enteras y positivas para 'Adult Male'
variation_adult_male = np.random.randint(low=0, high=5, size=len(df_usa))

# Generar variaciones aleatorias enteras y positivas para 'Adult Female'
variation_adult_female = np.random.randint(low=0, high=2, size=len(df_usa))

# Agregar variaciones a los datos existentes
df_usa['Adult Male'] += variation_adult_male
df_usa['Adult Female'] += variation_adult_female
```

```python
# Generar variaciones aleatorias enteras y positivas para 'Old Male'
variation_old_male = np.random.randint(low=0, high=4, size=len(df_usa))

# Generar variaciones aleatorias enteras y positivas para 'Old Female'
variation_old_female = np.random.randint(low=0, high=1, size=len(df_usa))

# Agregar variaciones a los datos existentes
df_usa['Old Male'] += variation_old_male
df_usa['Old Female'] += variation_old_female
```

```python
# Calculamos estadísticas como la media, mediana, etc.

# Restringir las fechas a 20 fechas
fechas_restringidas = df_usa['Date'].unique()[:10]

# Filtrar el DataFrame por las fechas restringidas
df_restringido = df_usa[df_usa['Date'].isin(fechas_restringidas)]

# Calcular estadísticas
media_ncc = df_restringido['No. of confirmed cases'].mean()
mediana_ncc = df_restringido['No. of confirmed cases'].median()
std_ncc = df_restringido['No. of confirmed cases'].std()

# Graficar los datos y las estadísticas calculadas
fig, ax = plt.subplots(figsize=(15, 6))  # Ajustar el tamaño de la figura

# Graficar los datos de NCC
ax.plot(df_restringido['Date'], df_restringido['No. of confirmed cases'], label='NCC')

# Añadir las estadísticas calculadas al gráfico
ax.axhline(media_ncc, color='red', linestyle='--', label='Media')
ax.axhline(mediana_ncc, color='green', linestyle='--', label='Mediana')
ax.axhline(std_ncc, color='blue', linestyle='--', label='Desviación Estándar')

# Añadir la leyenda y los ejes
ax.legend()
ax.set_xlabel('Fechas')
ax.set_ylabel('Número de casos confirmados')
ax.set_title('Estadísticas de NCC')

# Mostrar el gráfico
plt.savefig("ESTADISTICA NCC.PNG",dpi=300, format='png')
plt.show()
```

```python
######################Grafica pastel 1
# Crea un gráfico de pastel para comparar los datos de infectados y muertes
fig, ax = plt.subplots()
ax.pie([sum(NCC), sum(NCD)], labels=['No. of confirmed cases', 'No. of confirmed deaths'], autopct='%1.1f%%', startangle=90)
ax.axis('equal')  # Para asegurar que el gráfico sea un círculo y no una elipse
plt.title('Comparación de infectados y muertes en Estados Unidos de América')
plt.savefig("pie_ncc,nmc.png",dpi=300, format='png')
plt.show()
```

```python
# Obtener estadistica manualmente#####################GRAFICA UNO
total_confirmed_cases = sum(NCC)
total_confirmed_deaths = sum(NCD)
total_all_cases = total_confirmed_cases + total_confirmed_deaths

# Calcular los porcentajes
percentage_confirmed_cases = (total_confirmed_cases / total_all_cases) * 100
percentage_confirmed_deaths = (total_confirmed_deaths / total_all_cases) * 100

# Comparar con el gráfico
print("Porcentaje de casos confirmados (cálculo manual):", percentage_confirmed_cases)
print("Porcentaje de muertes confirmadas (cálculo manual):", percentage_confirmed_deaths)
```
