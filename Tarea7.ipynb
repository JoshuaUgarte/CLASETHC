```python
def es_primo(numero):# Define una función llamada es_primo que toma un argumento numero.
    if numero <= 1:# Comprueba si el número proporcionado es menor o igual a
        print("Error en el Número")# Imprime un mensaje de error si el número es menor o igual a 1.
        return False# para indicar que el número no es primo.
    for i in range(2, int(numero**0.5) + 1):# Inicia un bucle for que itera sobre los valores de i desde 2 hasta la raíz cuadrada entera de numero más 1.
        if numero % i == 0:# Verifica si el número es divisible por i sin dejar residuo.
            return False# Retorna False si encuentra un divisor distinto de 1 y el propio número, indicando que el número no es primo.
    return True# Retorna True si no se encuentra ningún divisor distinto de 1 y el propio número, lo cual indica que el número es primo.

while True:# Inicia un bucle while infinito.
    try:# Inicia un bloque try para manejar posibles errores durante la ejecución.
        entrada = int(input("Ingrese un número entero: "))# solicita al usuario que ingrese un número entero y lo convierte a un entero utilizando int(). El resultado se almacena en la variable entrada
        print(es_primo(entrada))# Llama a la función es_primo con el número ingresado por el usuario y muestra el resultado.
        break# Sale del bucle while si todo se ejecuta correctamente.
    except ValueError: # Captura una excepción de tipo ValueError, que ocurre si el usuario ingresa algo que no puede convertirse a un número entero.
        print("Entrada inválida. Por favor, ingrese un número entero.")# Imprime un mensaje de error si se produce un ValueError.
```

```python
# Importar la biblioteca matplotlib para visualización de datos
import matplotlib.pyplot as plt

# Funciones lambda para calcular promedio, varianza y desviación estándar
promedio = lambda arr: sum(arr) / len(arr)
varianza = lambda arr: sum((x - promedio(arr))**2 for x in arr) / len(arr)
desviacion_estandar = lambda arr: varianza(arr)**0.5

# Arreglo de números enteros entre 0 y 150
numeros = [i for i in range(151)]

# Filtrar números pares e impares
numeros_pares = list(filter(lambda x: x % 2 == 0, numeros))
numeros_impares = list(filter(lambda x: x % 2 != 0, numeros))

# Filtrar múltiplos de 2, 3 y 5
multiplos_dos = list(filter(lambda x: x % 2 == 0, numeros))
multiplos_tres = list(filter(lambda x: x % 3 == 0, numeros))
multiplos_cinco = list(filter(lambda x: x % 5 == 0, numeros))

# Calcular y mostrar resultados estadísticos
print("Promedio:", promedio(numeros))
print("Varianza:", varianza(numeros))
print("Desviación Estándar:", desviacion_estandar(numeros))

print("\nSuma de números pares:", sum(numeros_pares))
print("Suma de números impares:", sum(numeros_impares))

print("\nMúltiplos de 2:", multiplos_dos)
print("Múltiplos de 3:", multiplos_tres)
print("Múltiplos de 5:", multiplos_cinco)

# Visualización de resultados mediante gráficos

# Gráfico de pastel (pie chart) para los múltiplos de 2, 3 y 5
etiquetas = ['Multiplos de 2', 'Multiplos de 3', 'Multiplos de 5']
values = [len(multiplos_dos), len(multiplos_tres), len(multiplos_cinco)]

# Define una paleta de colores para el gráfico
colors = plt.cm.Set3(range(len(etiquetas)))

# Crea el gráfico de pastel
plt.pie(values, labels=etiquetas, colors=colors, autopct='%1.1f%%')

# Agrega un título al gráfico
plt.title('Multiplos de 2, 3 y 5')

# Muestra el gráfico
plt.show()

# Gráfico de barras (bar chart) para promedio, varianza y desviación estándar
etiquetas = ['Promedio', 'Varianza', 'Desviación Estándar']
valores = [promedio(numeros), varianza(numeros), desviacion_estandar(numeros)]

# Crea el gráfico de barras
plt.bar(etiquetas, valores)

# Agrega un título al gráfico
plt.title('Estadísticas')

# Muestra el gráfico de barras
plt.show()
```

```python
# Importar la biblioteca random para generar los números aleatorios
import random
# Importar la biblioteca matplotlib para visualizar los datos
import matplotlib.pyplot as plt

# para generar una lista de 150 números aleatorios entre 1 y 150
random_numbers = [random.randint(1, 151) for _ in range(151)]

# formula para calcular el promedio
mean = lambda data: sum(data) / len(data)
mean_value = mean(random_numbers)
print("Promedio:", mean_value)

#formula para calcular la varianza
variance = lambda data, mu: sum((x - mu) ** 2 for x in data) / len(data)
variance_value = variance(random_numbers, mean_value)
print("Varianza:", variance_value)

#formula para calcular la desviación estándar
std_deviation = lambda data, mu: (variance(data, mu))**0.5
std_deviation_value = std_deviation(random_numbers, mean_value)
print("Desviación estándar:", std_deviation_value)

#histogramas en uno solo
plt.figure(figsize=(15, 5))

# Histograma del promedio
plt.subplot(1, 4, 2)
plt.hist([mean_value] * len(random_numbers), color='lightgreen', edgecolor='black')
plt.title('promedio')

# Histograma de la varianza
plt.subplot(1, 4, 3)
plt.hist([variance_value] * len(random_numbers), color='lightcoral', edgecolor='black')
plt.title('Varianza')

# Histograma de la desviación estándar
plt.subplot(1, 4, 4)
plt.hist([std_deviation_value] * len(random_numbers), color='gold', edgecolor='black')
plt.title('Desviacion estandar')

#para la gráfica de barras
fig = plt.figure(figsize=(12, 10))
plt.bar(['Promedio', 'Varianza', 'Desviación Estándar'], [mean_value, variance_value, std_deviation_value], color=['lightgreen', 'lightcoral', 'gold'], alpha=0.7)
plt.title('Resultados de las operaciones')
plt.xlabel('Operaciones')
plt.ylabel('Valores')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.savefig("resultados_operaciones.png", dpi=300, format='png')
plt.show()

# para la gráfica de pastel
fig, ax = plt.subplots()
ax.pie([mean_value, variance_value, std_deviation_value], labels=['Promedio', 'Varianza', 'Desviación Estándar'], autopct='%1.1f%%', startangle=90, colors=['lightgreen', 'lightcoral', 'gold'])
ax.axis('equal')
plt.title('Distribución de resultados')
plt.show()
```

```python
#para immporta la biblioteca NumPy con el alias `np`.
import numpy as np

#para importa la biblioteca Matplotlib con el alias `plt`.
import matplotlib.pyplot as plt

# aque se define una función llamada `y` que representa la solución de la ecuación diferencial.
def y(x, C):
    return (x - 1) + C*np.exp(-x)

# aqui se crea un arreglo `x` que contiene 100 puntos equidistantes en el rango de -2 a 6.
x = np.linspace(-2, 6, 100)

# La sección siguiente usa un bucle `for` para graficar varias curvas con diferentes valores de `C`.
for C in np.linspace(0.1, 1, 5):
    # Crea una nueva figura con un tamaño de 8x5 pulgadas.
    plt.figure(figsize=(8, 5))

    # Grafica la curva utilizando la función `y` con el valor actual de `C` y etiqueta la curva con el valor de `C`.
    plt.plot(x, y(x, C), label=f"C={C}")

    # Agrega una leyenda a la gráfica.
    plt.legend()

    # Agrega un título a la gráfica que indica el valor actual de `C`.
    plt.title(f"Curva para C={C}")

    # Muestra la gráfica.
    plt.show()

# aqui se grafica la curva para `C = 0`.
plt.figure(figsize=(8, 5))
plt.plot(x, y(x, 0), label="C=0")
plt.legend()
plt.title("Curva para C=0")
plt.show()

# La sección siguiente usa otro bucle `for` para graficar curvas con valores de `C` negativos.
for C in np.linspace(-1, -0.1, 5):
    # Crea una nueva figura con un tamaño de 8x5 pulgadas.
    plt.figure(figsize=(8, 5))

    # Grafica la curva utilizando la función `y` con el valor actual de `C` y etiqueta la curva con el valor de `C`.
    plt.plot(x, y(x, C), label=f"C={C}")

    # Agrega una leyenda a la gráfica.
    plt.legend()

    # Agrega un título a la gráfica que indica el valor actual de `C`.
    plt.title(f"Curva para C={C}")

    # Muestra la gráfica.
    plt.show()

# La última sección crea una familia de curvas para diferentes valores de `C` en el rango de -1 a 1.
plt.figure(figsize=(8, 5))
for C in np.linspace(-1, 1, 5):
    # Grafica las curvas correspondientes.
    plt.plot(x, y(x, C), label=f"C={C}")

# Agrega una leyenda y un título a la última gráfica que representa la familia de curvas para la ecuación diferencial .
plt.legend()
plt.title("Familia de curvas para y=(x-1)+Ce^(-x)")
plt.show()
```

```python
import numpy as np #para immporta la biblioteca NumPy con el alias `np`.
import matplotlib.pyplot as plt #para importa la biblioteca Matplotlib con el alias `plt`.

# Define la función f(x)
def f(x):
    return (3*x + 7) / (x**2 - 4)

# Define el rango de valores para la gráfica
x = np.linspace(-10, 10, 1000)

# Grafica la función
plt.plot(x, f(x), '-k')

# Grafica las asíntotas horizontales
plt.axhline(y=0, color='red', linestyle='--')
plt.axhline(y=3, linestyle='--')

# Grafica la asíntota vertical
plt.axvline(x=2, linestyle='--')
plt.axvline(x=0, color='red', linestyle='--')

# Establece los límites en el eje "y"
plt.ylim(-10, 10)

# Añade una leyenda
plt.legend(['f(x)', 'y=0', 'y=3', 'x=2','X=0'], loc='upper left')

# Muestra la gráfica
plt.show()
```
